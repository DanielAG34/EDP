```python
# prompt: graficar funcion y=x∧2

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 5, 100)
y = x**2

plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y = x^2")
plt.title("Gráfica de $y = x^2$",color="Magenta")
plt.grid(True)
plt.show()
```

<font color="red">Grafica de la ecuacion cuadratica: $y=x^2$
