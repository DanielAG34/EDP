Grafique la función $f'(x)' = exp (- \dfrac{(x'-0.5)^2}{0.01})$ y suponga que esta representa una perturbación en una cuerda

```python
import numpy as np
import matplotlib.pyplot as plt

# Definición de la función
def f(x_prime):
    return np.exp(-(x_prime - 0.5)**2 / 0.01)

# Rango de valores para x'
x_prime = np.linspace(0, 1, 1000)

# Calculamos los valores de la función
y = f(x_prime)

# Configuración del gráfico
plt.figure(figsize=(8, 5))
plt.plot(x_prime, y, 'g-', linewidth=2)
plt.title('Perturbación inicial en la cuerda\n$f(x\') = \exp\\left(-\\frac{(x\' - 0.5)^2}{0.01}\\right)$')
plt.xlabel('Posición (x\')')
plt.ylabel('Amplitud de la perturbación')
plt.grid(True, linestyle='--', alpha=0.7)
plt.xlim(0, 1)
plt.ylim(0, 1.1)

# Destacamos el centro de la gaussiana
plt.axvline(x=0.5, color='b', linestyle='--', alpha=0.5)
plt.text(0.5, -0.1, 'x\' = 0.5', ha='center', va='top', color='r')

plt.show()
```

a$)$ Mostrar que al hacer un cambio de variable $x'= x-ct$ la perturbación se desplaza a la derecha.

b$)$ Mostrar que al hacer un cambio de variable $x'= x+ct$ la perturbación se desplaza a la izquierda.

c$)$ Graficar la superposicón de las ondas sw los anteriores incisos.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 1, 500)
c = 1
t = 0.2  # instante de tiempo
sigma = 0.01
x0 = 0.5


def f(xp):
    return np.exp(-((xp - x0)**2) / sigma)

# Inciso (a): f(x - ct)
fa = f(x - c*t)

# Inciso (b): f(x + ct)
fb = f(x + c*t)

# Inciso (c): superposición
fc = fa + fb

# Graficar
fig, axs = plt.subplots(3, 1, figsize=(8, 10))

axs[0].plot(x, fa)
axs[0].set_title('Inciso (a): f(x - ct) - desplazamiento a la derecha')
axs[0].grid()

axs[1].plot(x, fb)
axs[1].set_title('Inciso (b): f(x + ct) - desplazamiento a la izquierda')
axs[1].grid()

axs[2].plot(x, fc)
axs[2].set_title('Inciso (c): superposición f(x - ct) + f(x + ct)')
axs[2].grid()

plt.tight_layout()
plt.show()
```
